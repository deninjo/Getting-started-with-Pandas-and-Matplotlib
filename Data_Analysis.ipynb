```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_with_headers.csv')
data[0:5]#printing the first 5 rows
```

```python
Time = data['time']
Sensors = data.loc[:, 's1': 's4']#.loc is a slicing technique. ':' means all rows from column s1 to s4
print(Sensors[0:6])
```

```python
avg_row = np.mean(Sensors,1)#1 is for getting average for across rows
print(avg_row[0:4])
```

```python
avg_col = np.mean(Sensors,0)#0 is for getting average for across columns
print(avg_col[0:4])
```

```python
my_data = [Time, Sensors, avg_row]
result = pd.concat(my_data, axis=1)#axis=1 is stacking data items left and right to each other. axis=0 is stacking on top
print(result[0:5])
```

```python
result.to_csv('result.csv')#saving the result as a separate csv file
result.to_excel('result.xlsx')#saving the result as a separate excel file
```

```python
plt.plot(Time, Sensors['s1'], 'b-')
plt.plot(Time, avg_row, 'r.')
plt.legend(['Sensor 1', 'Average'])#A legend is an area describing the elements of the graph
plt.xlabel('Time (sec)')
plt.ylabel('Sensor Values')
```
